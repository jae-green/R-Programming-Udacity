### Gender, Tenure, & Labels

As you saw in the previous videos, you can get summary statistics and colors.  These are two new things you can start incorporating into analyzing data with R.

First, let's take a look at summary statistics by gender.

```r
fb = read.csv('./pseudo_facebook.tsv', sep='\t')
head(fb)
```

```r
table(fb$gender)
```

```r
by(fb$friend_count, fb$gender, summary)
```

You can now use your results to answer the questions posed in the earlier quiz.

You can also create a histogram of how long someone has been using facebook with the following code:

```r
library(ggplot2)
```

```r
ggplot(aes(x = tenure), data = fb) +
  geom_histogram(binwidth = 30, color = 'black', fill = '#099DD9')
```

If you wanted to measure in years instead of in days, you could run something like the following:

```r
ggplot(aes(x = tenure/365), data = fb) +
  geom_histogram(binwidth = .25, color = 'black', fill = '#F79420')
```

We can now add labels to these plots with the xlab and ylab arguments.

```r
ggplot(aes(x = tenure/365), data = fb) +
    geom_histogram(binwidth = .25, color = 'black', fill = '#F79420') +
    xlab('Number of years using Facebook') +
    ylab('Number of users in sample')
```
