### Friends

Next let's look at the number of friends users have on Facebook (another reminder that this is not real data).

````r
library(ggplot2)
fb = read.csv('./pseudo_facebook.tsv', sep='\t')
head(fb)
````

````r
ggplot(aes(x=friend_count), data=fb) +
  geom_histogram(binwidth=100) 
````

It's good to get in the habit of adding labels to the title and each axis of your visualizations.  For the same plot we can add labels as shown below.

````r
ggplot(aes(x=friend_count), data=fb) +
  geom_histogram(binwidth=100) + 
  ggtitle('Histogram of Number of Friends Per User') +
  labs(x = "Number of Friends")
````

You can now add limits to the plot to zoom in on the bulk of the distribution.  For this example, let's say up to 1,000 friends.

````r
ggplot(aes(x=friend_count), data=fb) +
  geom_histogram(binwidth=100) + 
  ggtitle('Histogram of Number of Friends Per User') +
  labs(x = "Number of Friends") + 
  scale_x_continuous(limits=c(0,1000))
````

Similar to the previous section, we can also use facets to look at the number of friends associated with each gender.  As a reminder, the two options for setting up a facet are:

```
facet_wrap(~variable_to_split_on)
facet_grid(rows_variable~columns~variable)
```

````r
ggplot(aes(x=friend_count), data=fb) +
  geom_histogram(binwidth=100) + 
  ggtitle('Histogram of Number of Friends Per User') +
  labs(x = "Number of Friends") + 
  scale_x_continuous(limits=c(0,1000)) + 
  facet_wrap(~gender)
````

Now let's omit the NA values.  Instead of using all of the data, we can remove the na values from the dataset associated with gender. 

You can see from the warning that you still have missing values associated with friend count.

````r
ggplot(aes(x=friend_count), data=subset(fb, !is.na(gender))) +
  geom_histogram(binwidth=100) + 
  ggtitle('Histogram of Number of Friends Per User') +
  labs(x = "Number of Friends") + 
  scale_x_continuous(limits=c(0,1000)) + 
  facet_wrap(~gender)
````
