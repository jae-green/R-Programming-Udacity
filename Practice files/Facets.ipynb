### Faceting

Faceting is a way to split our one plot into multiple plots based on another variable of interest.  

Let's consider we want to look at the distribution of birthdays per day, but we want to create a different plot for each month.  We can do this using faceting.  

Start by reading in the dataset and the ggplot2 library.

```r
library(ggplot2)

fb = read.csv('./pseudo_facebook.tsv', sep='\t')
```

We can use the grammar of graphics methods to create our plot in the following way:

```r
ggplot(aes(x = dob_day), data = fb) +
  geom_histogram(binwidth = 1) +
  scale_x_continuous(breaks = 1:31) +
  facet_wrap(~dob_month)
```

```r
head(fb)
```

We could do something similar but see if genders have differences in which days they are born on.

```r
ggplot(aes(x = dob_day), data = fb) +
  geom_histogram(binwidth = 1) +
  scale_x_continuous(breaks = 1:31) +
  facet_wrap(~gender) 
```

It can be difficult to compare the above directly.  One useful way to make comparisons between the two is to add `y` in the `aes` command set to `..density..`.  Below is an example:  

```r
ggplot(aes(x = dob_day, y = ..density..), data = fb) +
  geom_histogram(binwidth = 1) +
  scale_x_continuous(breaks = 1:31) +
  facet_wrap(~gender) 
```

We could use `facet_grid` to look at both gender and month at the same time.  Here is an example:

```r
ggplot(aes(x = dob_day, y = ..density..), data = fb) +
  geom_histogram(binwidth = 1) +
  scale_x_continuous(breaks = 1:31) +
  facet_grid(dob_month~gender) 
```

In the above, each row is a different month, while each column is a different gender.
